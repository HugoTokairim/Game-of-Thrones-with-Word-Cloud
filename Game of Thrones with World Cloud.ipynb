```python
df = pd.read_csv("D:\\NPower\\Projects\\Game of Thrones - WorldCloud\\game of thrones.csv")
df.head()
```

```python
# creating the text variable
text = " ".join(title for title in df.Title)
```

```python
# stopwords: ‘stop_words’ are those words that are commonly used in the English language such as ‘we’, ‘the’, ‘a’, ‘an’, etc. thus, we have to eliminate those words.
stopwords = set(STOPWORDS)
```

```python
wordcloud = WordCloud(stopwords = STOPWORDS).generate(text)
#Display the generated image
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
```

```python
#mask: Specify the shape of the word cloud image. By default, it takes a rectangle. As we know we have created a variable as a mask assign it to the mask parameter.
mask = np.array(Image.open("D:\\NPower\\Projects\\Game of Thrones - WorldCloud\\stark-sigil-winter-is-coming-png-clipart.jpg"))
plt.imshow(mask)
```

```python
# Creating word_cloud with text as argument in .generate() method
wordcloud = WordCloud(background_color = 'white',mask = mask, 
               width = 2048, height = 1080).generate(text)
```

```python
# font color matching the masked image
img_colors = ImageColorGenerator(mask)
wordcloud.recolor(color_func = img_colors)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
